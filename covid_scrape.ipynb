```python
#create an empty list
col= []
i= 0
```

```python
# for each row, store each first elemet(header) and an empty list
for t in tr_elem[0]:
    i+=1
    name= t.text_content()
    "%d:'%s'"
    print(i, name)
    col.append((name, []))
```

```python
# Since the first row is the header, the data is stored in the second row onwards
for j in range(1,len(tr_elem)):
    #T is out j'th row
    T= tr_elem[j]
    
    #if row is not the size of 5 the //tr data is not from our table
    if len(T) != 5:
        break
        
    # i is the index of our column
    i= 0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data= t.text_content()
        #Check if row is empty
        if i>0:
            #convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment of i for the next data column
        i+=1
```
